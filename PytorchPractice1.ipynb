**Saving and Loading Tensors**
In pytorch, the primary tools for saving tensors are torch.save and torch.load
These functions use Python's pickle utility to serialize objects to disk, typically using the .pt or .pth file extension

```python
# Saving a single tensor :
import torch
x = torch.tensor([[1,2,3],[4,5,6]])
torch.save(x, 'tensor.pt')

#Loading a single tensor :
y = torch.load('tensor.pt')
print(y)
```

```python
#Because Pytorch uses pickle, you can save collections like dictionaries or lists containing multiple tensors in a single file

#Saving multiple tensors
tensor_dict = {
    'weights' : torch.randn(2,3),
    'biases' : torch.zeros(2)
}
torch.save(tensor_dict, 'data.pt')

#Load back as a dictionary
loaded_data = torch.load('data.pt')
print("Loaded_data : \n",loaded_data['weights'])
print("\n")

x1 = loaded_data['weights']
x2 = loaded_data['biases']
print("x1 : ",x1)
print("x2 : ",x2)
print("\n")

# x1 and x2 are tensors of type float32
# Let us store them in different tensors of type int32
x3 = x1.type(torch.int32)
x4 = x2.type(torch.int32)

print("x3 : ",x3)
print("x4 : ",x4)
```
