**NUMPY ARRAYS**

```python
import numpy as np
#Creating numpy array using arange
my_numpy_array = np.arange(15)
my_numpy_array
```

```python
#Creating numpy array using numpy.array()
my_numpy_array1 = np.array([[1,2,3],[4,5,6]])
my_numpy_array1
```

```python
#Creating numpy array using random.rand()
my_numpy_array2 = np.random.rand(3,4,5)
print(my_numpy_array2, my_numpy_array2.dtype)
```

**TENSORS**

```python
import torch
#Creating tensor using torch.randn()
tensor_1 = torch.randn(3,4)
print(tensor_1, tensor_1.dtype)
```

```python
#Creating tensor from numpy array
tensor_2 = torch.tensor(my_numpy_array2)
tensor_2
```

```python
#Creating tensors with all ones or zeroes.
tensor_4 = torch.ones([5,4,3], dtype=torch.int32)
tensor_5 = torch.zeros([7,3,2], dtype=torch.int32)
print(tensor_4)
print(tensor_5)
```

```python
print("Shape of tensor_4 = ", tensor_4.shape,"Shape of tensor_5 = ",  tensor_5.shape)
print("Layout of tensor_4 = ", tensor_4.layout,"Layout of tensor_5 = ",  tensor_5.layout)
print("Stride of tensor_4 = ", tensor_4.stride(), "Stride of tensor_5 = ", tensor_5.stride())

```

```python
tensor_6 = torch.randn(2,3)
tensor_7 = tensor_6.t()
print(tensor_6, "; shape of tensor_6 = ", tensor_6.shape, "; stride of tensor_6 = ", tensor_6.stride())
print(tensor_7, "; shape of tensor_7 = ", tensor_7.shape, "; stride of tensor_7 = ", tensor_7.stride())
```
